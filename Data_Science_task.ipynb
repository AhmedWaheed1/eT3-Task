```python
import pandas
dataset = pandas.read_csv('/content/drive/MyDrive/co/drinkMenu.csv')
```

```python
# Check null values on each column
dataset.isnull().sum()
```

```python
# The check show than only the last column has only a null value ,then impute it (replace by 0 for example)
dataset['Caffeine (mg)'].fillna( value='0' , inplace=True )
```

```python
# Remove duplicates based on the first 3 columns
dataset.drop_duplicates( subset= ["Beverage_category","Beverage","Beverage_prep"] , keep='last' , inplace=True)
```

```python
# Q1 visualize the Highest Calories -> Signature Espresso Drinks = 510
import matplotlib.pyplot as plt

plt.figure( figsize= (30,5) )
plt.plot( dataset['Beverage_category'] , dataset['Calories'] , 'o' )
plt.annotate( 'Highest Calories' , xy=(2,510) , xytext=(2.5,550) , arrowprops=dict(facecolor='black') , fontsize=30 , color='red' )

plt.xlabel( "Beverage_category" , fontsize=25 , color='blue' )
plt.ylabel( "Calories"          , fontsize=25 , color='blue' )
```
